```python
Lucknow = data[data['City']=='Lucknow']
Lucknow
```

```python
for column in Lucknow[['PM2.5', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']]:
    print(column)
    count = Lucknow[column].isna().sum()
    print((count/2009)*100)
```

```python
Lucknow = Lucknow.drop(columns = ['PM10','NH3','Xylene'])
Lucknow
```

```python
for column in Lucknow[['PM2.5', 'NO', 'NO2', 'NOx', 'CO', 'SO2',
       'O3', 'Benzene', 'Toluene']]:
  colMean = Lucknow[column].mean(skipna = True)
  Lucknow[column].fillna(colMean)
```

```python
print(Lucknow['AQI'].mean(skipna = True))
Lucknow['AQI'].fillna(218, inplace = True)
```

```python
Lucknow['AQI_Bucket'].value_counts()
```

```python
Lucknow['AQI_Bucket'].fillna('Moderate', inplace = True)
```

```python
Lucknow
```

```python
Lucknow['Date'] = pd.to_datetime(Lucknow['Date'])
```

```python
Lucknow['City'] = Lucknow['City'].str.lower()
Lucknow['AQI_Bucket'] = Lucknow['AQI_Bucket'].str.lower()
```

```python
Lucknow['month'] = pd.DatetimeIndex(Lucknow['Date']).month
Lucknow['year'] = pd.DatetimeIndex(Lucknow['Date']).year
Lucknow.head()
Lucknow
```

```python
Lucknow = Lucknow [ Lucknow ["year"]>=2017]
Lucknow
```

```python
df_corr = Lucknow.corr()
df_corr = df_corr.drop(['year','month'],axis = 1)
df_corr = df_corr.drop(['year','month'],axis = 0)
plt.figure(figsize=(20,8))
ax = sns.heatmap( df_corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True, annot = True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
```

```python
lucknow_2020 = Lucknow[Lucknow['year'] == 2020] 
lucknow_2020
```

```python
plt.figure(figsize=(10,8))
plt.plot(lucknow_2020['Date'],lucknow_2020['AQI'])
plt.xlabel("Date")
plt.ylabel("AQI")
plt.title("AQI vs Date graph for the year 2020 Location : Lucknow")
plt.legend(["Lucknow"])
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (20,8))
ax1.title.set_text("Lucknow 2020 - NO2 conc.")
lucknow_2020.groupby(lucknow_2020['month']).mean().plot(y = ['NO2'],kind = 'bar',ax =ax1)
ax2.title.set_text("Lucknow 2020 - CO conc.")
lucknow_2020.groupby(lucknow_2020['month']).mean().plot(y = ['CO'],kind = 'bar',ax = ax2)
ax3.title.set_text("Lucknow 2020 - SO2 conc.")
lucknow_2020.groupby(lucknow_2020['month']).mean().plot(y = ['SO2'],kind = 'bar',ax = ax3)

plt.plot()
```
